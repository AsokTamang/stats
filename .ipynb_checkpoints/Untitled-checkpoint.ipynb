```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
import seaborn as sns
```

```python
mu = 10
sigma = 5
gaussian_distribution = np.random.normal(mu,sigma,10000)
gaussian_distribution
```

```python
sns.histplot(gaussian_distribution,stat='density')
plt.show()
```

```python
def sample_means(data,sample_size):
    means = []
    for i in range(10000):
        sample = np.random.choice(data,size = sample_size)  #taking an array of sample data of size 5 currently
        mean_value = np.mean(sample)
        means.append(mean_value)
    return np.array(means)
gaussian_sample_means = sample_means(gaussian_distribution,5)    
sns.histplot(gaussian_sample_means,stat = 'density')
```

```python
# Compute estimated mu
mu_sample_means = mu

# Compute estimated sigma
# 5 is being used because you used a sample size of 5
sigma_sample_means = sigma / np.sqrt(5)


x_range = np.linspace(min(gaussian_sample_means), max(gaussian_sample_means), 100)

sns.histplot(gaussian_sample_means, stat="density")
plt.plot(
    x_range,
    norm.pdf(x_range, loc=mu_sample_means, scale=sigma_sample_means),
    color="black",
)
plt.show()
```
